# Look at what hight is required for 2nd order mode

```python
import numpy as np
import tidy3d as td
import matplotlib.pyplot as plt
import tidy3d.web as web
from tidy3d.plugins import waveguide
from tidy3d.plugins.mode.web import run as run_mode_solver
```

```python
InP = td.Medium(permittivity=3.13**2)
Air = td.Medium(permittivity=1.0**2)

#define the waveguide 
strip = waveguide.RectangularDielectric(
    wavelength=1.55,
    core_width=0.55,
    core_thickness=0.29,
    core_medium=InP,
    clad_medium=Air,
    mode_spec=td.ModeSpec(num_modes=4)
)
fig, ax = plt.subplots(4,3,figsize=(10,10))
strip.plot_field("Ex", mode_index=0, ax=ax[0,0])
strip.plot_field("Ey", mode_index=0, ax=ax[0,1])
strip.plot_field("Ez", mode_index=0, ax=ax[0,2])
strip.plot_field("Ex", mode_index=1, ax=ax[1,0])
strip.plot_field("Ey", mode_index=1, ax=ax[1,1])
strip.plot_field("Ez", mode_index=1, ax=ax[1,2])
strip.plot_field("Ex", mode_index=2, ax=ax[2,0])
strip.plot_field("Ey", mode_index=2, ax=ax[2,1])
strip.plot_field("Ez", mode_index=2, ax=ax[2,2])
strip.plot_field("Ex", mode_index=3, ax=ax[3,0])
strip.plot_field("Ey", mode_index=3, ax=ax[3,1])
strip.plot_field("Ez", mode_index=3, ax=ax[3,2])
plt.show()
```
