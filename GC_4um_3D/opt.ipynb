# 3D optimization of grating coupler

```python
import autograd.numpy as np 
import matplotlib.pyplot as plt

import tidy3d.web as web
import pickle

from main import (make_sim, get_coupling_efficiency, projection_builder, run_adam,
                  r0, n_wl, wl_range, apodized_to_widths)
N_teeth = 9
```

```python
with open('data/2d_opt/first30.pkl', 'rb') as f:
    data = pickle.load(f)

widths = data['params'][:-1]
r0 = data['params'][-1]+2.5
```

```python
def objective(params,projection=lambda x: x,path='/Users/dominic/Desktop/fdtd/testing/PhotonicComponents/GC_4um_3d/data/tidy3d_output/temp.hdf5'):
    #pull out the parameters
    params_proj = projection(params)
    widths = params_proj[:N_teeth*2+2]
    r0 = params_proj[N_teeth*2+2]
    
    #run simulation
    sim = make_sim(widths, N_teeth=N_teeth, r0=r0, min_steps_per_wvl=20)
    sim_data = web.run(sim, task_name="GC4um_2D_SIO", folder_name="dominic", verbose=False, path=path)
    return get_coupling_efficiency(sim_data)[n_wl//2] 
```

```python
params = np.array([*widths,r0])

project, inverse_project = projection_builder()

history, opt_state = run_adam(params, project, inverse_project, objective, num_steps=30, learning_rate=0.02, verbose=True)
```
